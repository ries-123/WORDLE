```python
# link to github repository: https://github.com/ries-123/WORDLE 
## importing libraries. random is used as a random number generator to randomly select a word from our list of astro words. mathplotlib is used to create our graph for the winning statistics to show how many guesses you used each attempt.
import numpy as np
import random
import matplotlib.pyplot as plt

## list of astronomy words pulled from glossaries and other sites; these will be the words used in the wordle game. the words are listed from easiest to most difficult, which will later be set to the easy, hard, and ultrakill levels. 
astro_words = [
    "Space", "Earth", "Lunar", "Solar",
    "Orbit", "Light", "Night", "Comet", 
    "Radio", "Pluto", "Alpha", "Venus",
    "Phase", "Flare",
    "Mount", "Aries", "Dwarf", "Orion",
    "Titan", "Bulge", "Libra", "Virgo",
    "Epoch", "Thebe", "Ceres", "Umbra", 
    "Helix", "Deneb", "Nadir", "Apsis",
    ]
## word selection moved after difficulty. these are our variables we are going to be using later on in other parts of the code
guesses_left = 6 # of guesses left, will be printed after each guess and 1 will be subtracted each time so it counts down
guesses_so_far = 0 # guesses # so far, will be printed after each guess and 1 will be added each time so it counts up
appended_guess = [] #list of words guessed, will be printed after each guess - list
n = 0 #this is the thing that will count up to test each letter (see below)
appended_letters = [] #list of letters that have been guessed so far, will be printed after each guess to show guessed letters - list
guessed_letter_colors = {} # maps letter to its most recent color code - dictionary
colored_letters = [] #list of letters with colors, will be printed after each guess - list

#variables that will be used for the graph and statistics. one will be added to the correct guess_on_# after you correctly guess the word on that guess
guess_on_one = 0 # if guessed on first attempt
guess_on_two = 0 # if guessed on second attempt
guess_on_three = 0 #if guessed on third attempt
guess_on_four = 0 # if guessed on fourth attempt
guess_on_five = 0 # if guessed on fifth attempt
guess_on_six = 0 # if guessed on sixth attempt
loss_counter = 0 # if not guessed within six attempts, add one loss
play = "yes"

print("AstWordle: Astronomy meets Wordle. ")
#game loop, makes it so that all data and statistics will be stored with each time you play the game and so that we are able to update the counters and graph each time you play
while play=="yes":
    while True:  # loop for difficulties
        difficulty = input("choose your difficulty: easy, hard or ULTRAkill!!!: ").lower() # where you will enter the difficulty you want to play on, makes sure that whatever you type in is lowercase and will match the three difficulties
        if difficulty == "easy":
            astro_words_difficulty = astro_words[0:9] # easy words are the first 10 words
            break
        elif difficulty == "hard":
            astro_words_difficulty = astro_words[10:19] # hard words are the second 10 words
            break
        elif difficulty == "ultrakill":
            astro_words_difficulty = astro_words[20:29] # ultrakill words are the last 10 words
            break
        else:
            print("Please enter easy, hard, or ultrakill.") # what will be printed if you enter anything that isnt the three difficulties and will let you enter a difficulty again
    # Now select the word from the correct difficulty
    word = random.choice(astro_words_difficulty).upper() # makes it so that the word chosen is a random word from the astro word list based on the difficulty you chose, and makes it uppercase for cool wordle format
    word_letters = list(word) # turns word into a list of individual letters to be called back to later for colors based on if they're correct or in the right spot
    print("you have 6 tries. after 3, you will be offered a hint.") 
    ##loops until user inputs the right type of stuff

    while guesses_left > 0: # loop for stuff while playing the game. stops the game from playing after the guesses run out and you are asked to play again
        #-------this lovely chunk of text will be for the llm hint thing-------------
        # We will use this to suppress some warnings that are not important
        import warnings
        # Suppress specific Pydantic warnings that clutter the output
        warnings.filterwarnings("ignore", category=UserWarning, module="pydantic")
        # We will use dotenv to read the .env file
        from dotenv import load_dotenv
        load_dotenv()
        import litellm
        import os
        from sklearn.cluster import KMeans
        custom_api_base = "https://litellmproxy.osu-ai.org/" # this is the api that we will be using
        astro1221_key = os.getenv("ASTRO1221_API_KEY") # this gets the key for the llm from the .env file
        warnings.filterwarnings("ignore", category=UserWarning, module="pydantic")
        #Removing unnecessary warnings."

        if guesses_left == 3: # checks when you have three guesses left; this is where we will start putting the hint part
            hint = input("do you want a hint? yes/no: ").lower() # where you will put if you want a hint or not; makes your response lowercase
            if hint == "yes": 
                    response = litellm.completion( 
                        model="openai/GPT-4.1-mini", # model for llm
                        messages=[
                            {"role": "user", "content": f"The word is {word}. Give me a hint that will help me guess the word in the context of astronomy in less than 100 characters."} #hint for the word, makes sure the hint is for the astronomy word
                        ], 
                        max_tokens=100, # to avoid it from being too slow or giving too long of a hint
                        api_base=custom_api_base, # base for api
                        api_key=astro1221_key,
                    )
                    print(response.choices[0].message.content) # prints the response generated by the llm

        while True: 
            guess = input("Enter a 5-letter word: ").upper() # where you will enter your five letter word to guess what the correct word could be
            if len(guess) != 5: # will make you guess again if the length of your guessed word is longer than five letters
                print("Please enter exactly 5 letters.")
                continue
            if not guess.isalpha(): # will make you guess again if your guess has anything that isn't a letter
                print("Please enter only letters.")
                continue
            break


#win conditions, helps keep track of wins per guess stats
        if guess == word and guesses_so_far == 0: # if you guess the word right on the first try it will print this
            guess_on_one += 1
            print(f"\x1b[46m congratulations you did it!! the word was {word}. you guessed it on guess #{guesses_so_far + 1}\x1b[0m")
            break
        elif guess == word and guesses_so_far == 1: # if you guess the word right on the second try it will print this
            guess_on_two += 1
            print(f"\x1b[46m congratulations you did it!! the word was {word}. you guessed it on guess #{guesses_so_far + 1}\x1b[0m")
            break
        elif guess == word and guesses_so_far == 2: # if you guess the word right on the third try it will print this
            guess_on_three += 1
            print(f"\x1b[46m congratulations you did it!! the word was {word}. you guessed it on guess #{guesses_so_far + 1}\x1b[0m")
            break
        elif guess == word and guesses_so_far == 3: # if you guess the word right on the fourth try it will print this
            guess_on_four += 1
            print(f"\x1b[46m congratulations you did it!! the word was {word}. you guessed it on guess #{guesses_so_far + 1}\x1b[0m")
            break
        elif guess == word and guesses_so_far == 4: # if you guess the word right on the fifth try it will print this
            guess_on_five += 1
            print(f"\x1b[46m congratulations you did it!! the word was {word}. you guessed it on guess #{guesses_so_far + 1}\x1b[0m")
            break
        elif guess == word and guesses_so_far == 5: # if you guess the word right on the sixth try it will print this
            guess_on_six += 1
            print(f"\x1b[46m congratulations you did it!! the word was {word}. you guessed it on guess #{guesses_so_far + 1}\x1b[0m")
            break

        
        
        guess_letters = list(guess) #turns guess into letters, separates them into different strings to test
        

        for n in range(5): # makes what happens under happen for each letter in the word you guess
            letter = guess_letters[n]
            # Determine color code for each letter in this guess, and making the text of the letter itself white
            if guess_letters[n] == word_letters[n]:
                color = '\x1b[102m'  # makes the letter background green if the letter in the word you guessed in the correct spot
                colored_letters.append(f"\x1b[97m\x1b[102m{letter}\x1b[0m")
                print(f"in {guess}, \x1b[97m\x1b[102m{letter}\x1b[0m is correct and in the right position")
            elif letter in word_letters:
                color = '\x1b[103m'  # makes the letter background yellow if the letter in the word you guessed is in the word but not in the correct spot
                colored_letters.append(f"\x1b[97m\x1b[103m{letter}\x1b[0m")
                print(f"in {guess}, \x1b[97m\x1b[103m{letter}\x1b[0m is correct but in the wrong position")
            elif letter not in word_letters:
                color = '\x1b[100m'  # makes the letter background gray if the letter in the word you guessed is not in the word at all
                colored_letters.append(f"\x1b[97m\x1b[100m{letter}\x1b[0m")
                print(f"in {guess}, \x1b[97m\x1b[100m{letter}\x1b[0m is not in the word")
            else:
                color = ''
            # Always update the color for the letter to the most recent; adds the letters with their color to our appended letters that show all guessed letters when you are trying to make your next guess. it also prevents duplicates of letters
            guessed_letter_colors[letter] = color
            if letter not in appended_letters:
                appended_letters.append(letter)

        # Turns list into a string for later
        ("".join(colored_letters))
        n = 0

        guesses_left -= 1 # subtracts one from your guesses left after you have made a guess
        guesses_so_far += 1 #  adds one to your guesses so far after you have made a guess


        # Print all unique letters guessed so far (sorted, with color)
        print("letters guessed so far:") 
        sorted_letters = sorted(appended_letters) # adds new appended letters from guesses and sorts them alphabetically
        colored_guessed = [f"{guessed_letter_colors[l]}{l}\x1b[0m" for l in sorted_letters] 
        print(" ".join(colored_guessed)) 
        print(f"you have {guesses_left} guesses left")

        if guesses_left <= 0: # what happens when you run out of guesses. it will print this and then end the game--will still ask you later if you want to play again
            print(f"\x1b[31m*extremely loud incorrect buzzer*\x1b[0m] nah bru it was {word}")
            loss_counter += 1
            break

    #ai gives astro related explanations
    response = litellm.completion(
        model="openai/GPT-4.1-mini",
        messages=[
            {"role": "user", "content": f"The word is {word}. define in less than 300 characters and explain how this word relates to astronomy."} # prompt for the explanation of the guessed astronomy word that the llm will generate
        ],
        max_tokens=300,
        api_base=custom_api_base,
        api_key=astro1221_key,
    )
    print(response.choices[0].message.content) # prints the llm's response
    
    guess_graph_x = ['1', '2', '3', '4', '5', '6'] #x axis for graph, will be used to show how many guesses it took to win
    guess_graph_y = [guess_on_one, guess_on_two, guess_on_three, guess_on_four, guess_on_five, guess_on_six] # y-axis for graph, based on number of times you have guessed the correct word and how many guesses that took
    plt.bar(guess_graph_x, guess_graph_y, color='green') #makes the graph
    plt.xlabel('Guess Number') #x axis label    
    plt.ylabel('Number of Wins') #y axis label
    plt.title("AstWordle Guess Distribution") #title of graph
    plt.show() #shows the graph

    #prints wins to loss ratio
    win_total = guess_on_one + guess_on_two + guess_on_three + guess_on_four + guess_on_five + guess_on_six
    print(f"wins: {win_total}, losses: {loss_counter}, win/loss ratio: {win_total / loss_counter if loss_counter > 0 else 'N/A'}")

    #resetting variables for next game
    guesses_left = 6
    guesses_so_far = 0 
    appended_guess = [] 
    appended_letters = [] 
    guessed_letter_colors = {} 
    colored_letters = [] 

     #asking user to play again
    while True:
        play=input("play again? yes/no: ").lower()
        
        if play == "yes":
            break
        if play == "no":
            quit()
            break
        else:
            break
```
